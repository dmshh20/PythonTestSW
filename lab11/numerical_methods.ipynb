## 4) Додатково: Broadcasting та reshape

Поясніть що таке broadcasting і продемонструйте приклад. Спробуйте використовувати reshape для зміни розмірностей.

```python
# Broadcasting example
v = np.array([1, 2, 3])        # shape (3,)
M = np.array([[10, 20, 30],    # shape (2,3)
              [40, 50, 60]])
print('v shape:', v.shape)
print('M shape:', M.shape)
print('M + v -> broadcasting adds v to each row of M:\n', M + v)

# reshape example
d = np.arange(6)
print('\nd original:', d, 'shape', d.shape)
d2 = d.reshape((2,3))
print('d reshaped to (2,3):\n', d2)
```
