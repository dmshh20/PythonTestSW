## Line plot — приклад та homework1

Побудуємо синусіду та косинусіду. Homework1: поясніть, чому лінії плавні, як залежить вигляд від кількості точок.

```python
def homework1():
    print('>>>>> Homework1: поясніть формулу генерації точок та вплив кількості точок на гладкість графіка')

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(8,4))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)', linestyle='--')
plt.title('Line plot: sin and cos')
plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.grid(True)
plt.savefig('figure_line.png')
plt.show()

homework1()
```

## Scatter plot та histogram — приклади та homework2

Homework2: створіть розсіювання з шумом та гістограму розподілу випадкових значень; поясніть різницю.

```python
def homework2():
    print('>>>>> Homework2: додайте пояснення коли використовувати scatter vs histogram')

np.random.seed(0)
x = np.random.normal(loc=0.0, scale=1.0, size=200)
y = 0.5 * x + np.random.normal(scale=0.5, size=200)
plt.figure(figsize=(6,4))
plt.scatter(x, y, alpha=0.7)
plt.title('Scatter plot with noise')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
plt.hist(x, bins=20, alpha=0.8)
plt.title('Histogram of x')
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()

homework2()
```

## Subplots та порівняння — homework3

Homework3: створіть 2x2 subplot з різними функціями (sin, cos, tan, exp) та поясніть масштабування осей.

```python
def homework3():
    print('>>>>> Homework3: створіть 2x2 subplot і поясніть відмінності масштабів осей')

x = np.linspace(0.1, 2*np.pi, 200)
fig, axes = plt.subplots(2, 2, figsize=(10,8))
axes[0,0].plot(x, np.sin(x)); axes[0,0].set_title('sin')
axes[0,1].plot(x, np.cos(x), color='orange'); axes[0,1].set_title('cos')
axes[1,0].plot(x, np.tan(x), color='green'); axes[1,0].set_ylim(-10,10); axes[1,0].set_title('tan (clipped)')
axes[1,1].plot(x, np.exp(-x), color='red'); axes[1,1].set_title('exp(-x)')
for ax in axes.flatten():
    ax.grid(True)
plt.tight_layout()
plt.savefig('figure_subplots.png')
plt.show()

homework3()
```

## Збереження фігур та налаштування стилів — homework4

Homework4: збережіть фігуру у форматі PNG та змініть стиль matplotlib (наприклад 'seaborn').

```python
def homework4():
    print('>>>>> Homework4: застосуйте інший стиль і збережіть фігуру як PNG')

plt.style.use('seaborn')
x = np.linspace(0, 4*np.pi, 300)
y = np.sin(x) * np.exp(-x/10)
plt.figure(figsize=(8,4))
plt.plot(x, y)
plt.title('Damped sine (seaborn style)')
plt.xlabel('x')
plt.ylabel('amplitude')
plt.grid(True)
plt.savefig('figure_homework4.png')
plt.show()

homework4()
```
